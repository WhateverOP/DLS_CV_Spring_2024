# Метрика

Перед решением любой задачи важно понимать, как будет оцениваться ваше решение. В данном случае мы используем стандартную для задачи классификации метрику ROC-AUC. Ее можно вычислить, используя только предсказанные вероятности и истинные классы без конкретного порога классификации + она раотает даже если классы в данных сильно несбалансированны (примеров одного класса в десятки раз больше примеров длугого). Именно поэтому она очень удобна для соревнований.

Посчитать ее легко:


```python
from sklearn.metrics import roc_auc_score, roc_curve

y_true = [
    0,
    1,
    1,
    0,
    1
]

y_predictions = [
    0.1,
    0.9,
    0.4,
    0.6,
    0.61
]

roc_auc_score(y_true, y_predictions)
```

# ------------------------------------- LogReg ------------------------------------

```python
Log_reg_clf = LogisticRegression()

Log_reg_clf.get_params()
```

```python
Log_reg_param_grid = {
    'C': [1000, 100, 10, 1, 0.1, 0.01, 0.001],
    'penalty': ['l1', 'l2'],
    'max_iter': [10, 100, 500],
    'tol': [1e-5,1e-4,1e-3],
    'class_weight': [None,'balanced'],
    'solver': ['lbfgs','saga']
}

search = GridSearchCV(Log_reg_clf, Log_reg_param_grid, n_jobs=-1, cv=5, refit=True, scoring='roc_auc')
```

```python
search.fit(X_train_scaled, y_train)
```

```python
print(search.best_params_)
```

```python
best_Log_reg_clf = search.best_estimator_
```

```python
y_train_Log_reg_preds = best_Log_reg_clf.predict(X_train_scaled)
y_test_Log_reg_preds  = best_Log_reg_clf.predict(X_test_scaled)

y_train_Log_reg_prob_preds = best_Log_reg_clf.predict_proba(X_train_scaled)[:, 1]
y_test_Log_reg_prob_preds  = best_Log_reg_clf.predict_proba(X_test_scaled)[:, 1]
```

```python
y_train_Log_reg_preds_threshold = (y_train_Log_reg_prob_preds >= 0.5).astype(int) 
y_test_Log_reg_preds_threshold = (y_test_Log_reg_prob_preds >= 0.5).astype(int) 

train_Log_reg_roc_auc_score_threshold = roc_auc_score(y_train, y_train_Log_reg_preds_threshold)
test_Log_reg_roc_auc_score_threshold  = roc_auc_score(y_test, y_test_Log_reg_preds_threshold)
print('train_Log_reg_roc_auc_score_threshold =', train_Log_reg_roc_auc_score_threshold)
print('test_Log_reg_roc_auc_score_threshold =', test_Log_reg_roc_auc_score_threshold)
```

```python
y_train_Log_reg_preds_threshold = (y_train_Log_reg_prob_preds >= 0.1).astype(int) 
y_test_Log_reg_preds_threshold = (y_test_Log_reg_prob_preds >= 0.1).astype(int) 

train_Log_reg_roc_auc_score_threshold = roc_auc_score(y_train, y_train_Log_reg_preds_threshold)
test_Log_reg_roc_auc_score_threshold  = roc_auc_score(y_test, y_test_Log_reg_preds_threshold)
print('train_Log_reg_roc_auc_score_threshold =', train_Log_reg_roc_auc_score_threshold)
print('test_Log_reg_roc_auc_score_threshold =', test_Log_reg_roc_auc_score_threshold)
```

```python
train_auc = roc_auc_score(y_train, y_train_Log_reg_prob_preds)
test_auc = roc_auc_score(y_test, y_test_Log_reg_prob_preds)

plt.figure(figsize=(16,8))
plt.plot(*roc_curve(y_train, y_train_Log_reg_prob_preds)[:2], label='train AUC={:.4f}'.format(train_auc))
plt.plot(*roc_curve(y_test, y_test_Log_reg_prob_preds)[:2], label='test AUC={:.4f}'.format(test_auc))
legend_box = plt.legend(fontsize='large', framealpha=1).get_frame()
legend_box.set_facecolor("white")
legend_box.set_edgecolor("black")
plt.xlabel('False Positive Rate')
plt.ylabel('True Positive Rate')
plt.plot(np.linspace(0,1,100), np.linspace(0,1,100))
plt.show()
```

```python
class XGBoostIterativeOptimizer:
    def __init__(self, cv=5, n_iterations=3, random_state=42):
        """
        Initialize the XGBoostIterativeOptimizer.

        Parameters:
        - cv: Number of cross-validation folds for model evaluation.
        - n_iterations: Number of iterations for hyperparameter tuning.
        - random_state: Random state for reproducibility.
        """
        self.cv = cv
        self.n_iterations = n_iterations
        self.random_state = random_state
        self.best_params = {}

    def optimize(self, X, y):
#         best_params = {}

        for iteration in range(self.n_iterations):
            print(f"Iteration {iteration + 1}: Tuning Booster Parameters")
            booster_params = self.tune_booster_params(X, y)
            self.best_params.update(booster_params)

            print(f"Iteration {iteration + 1}: Tuning Tree-Specific Parameters")
            tree_params = self.tune_tree_params(X, y)
            self.best_params.update(tree_params)

            print(f"Iteration {iteration + 1}: Tuning Regularization Parameters")
            reg_params = self.tune_regularization_params(X, y)
            self.best_params.update(reg_params)

        return self.best_params

    def tune_booster_params(self, X, y):
        param_grid = {
            'max_depth': [3, 4, 5],
            'learning_rate': [0.1, 0.01, 0.001],
            'n_estimators': [100, 200, 300, 500]
        }

        optimizer = RandomizedSearchCV(
            xgboost.XGBClassifier(),
            param_distributions=param_grid,
            n_iter=10,
            cv=self.cv,
            n_jobs=-1,
            scoring='roc_auc',
            random_state=self.random_state
        )

        optimizer.fit(X, y)
        return optimizer.best_params_

    def tune_tree_params(self, X, y):
        param_grid = {
            'subsample': [0.8, 0.9, 1.0],
            'colsample_bytree': [0.8, 0.9, 1.0]
        }

        optimizer = RandomizedSearchCV(
            xgboost.XGBClassifier(**self.best_params),  # Use the best booster params from the previous iteration
            param_distributions=param_grid,
            n_iter=10,
            cv=self.cv,
            n_jobs=-1,
            scoring='roc_auc',
            random_state=self.random_state
        )

        optimizer.fit(X, y)
        return optimizer.best_params_

    def tune_regularization_params(self, X, y):
        param_grid = {
            'reg_alpha': [0, 0.001, 0.01, 0.1, 1],
            'reg_lambda': [0, 0.001, 0.01, 0.1, 1]
        }

        optimizer = RandomizedSearchCV(
            xgboost.XGBClassifier(**self.best_params),  # Use the best booster and tree params
            param_distributions=param_grid,
            n_iter=10,
            cv=self.cv,
            n_jobs=-1,
            scoring='roc_auc',
            random_state=self.random_state
        )

        optimizer.fit(X, y)
        return optimizer.best_params_
```

```python
def objective(trial):
    dtrain = xgboost.DMatrix(X_train_scaled, label=y_train)
    dvalid = xgboost.DMatrix(X_test_scaled, label=y_test)

    param = {
        "verbosity": 0,
        "objective": "binary:logistic",
        # use exact for small dataset.
        "tree_method": "exact",
        # defines booster, gblinear for linear functions.
        "booster": trial.suggest_categorical("booster", ["gbtree", "gblinear", "dart"]),
        # L2 regularization weight.
        "lambda": trial.suggest_float("lambda", 1e-8, 1.0, log=True),
        # L1 regularization weight.
        "alpha": trial.suggest_float("alpha", 1e-8, 1.0, log=True),
        # sampling ratio for training data.
        "subsample": trial.suggest_float("subsample", 0.2, 1.0),
        # sampling according to each tree.
        "colsample_bytree": trial.suggest_float("colsample_bytree", 0.2, 1.0),
    }

    if param["booster"] in ["gbtree", "dart"]:
        # maximum depth of the tree, signifies complexity of the tree.
        param["max_depth"] = trial.suggest_int("max_depth", 3, 9, step=2)
        # minimum child weight, larger the term more conservative the tree.
        param["min_child_weight"] = trial.suggest_int("min_child_weight", 2, 10)
        param["eta"] = trial.suggest_float("eta", 1e-8, 1.0, log=True)
        # defines how selective algorithm is.
        param["gamma"] = trial.suggest_float("gamma", 1e-8, 1.0, log=True)
        param["grow_policy"] = trial.suggest_categorical("grow_policy", ["depthwise", "lossguide"])

    if param["booster"] == "dart":
        param["sample_type"] = trial.suggest_categorical("sample_type", ["uniform", "weighted"])
        param["normalize_type"] = trial.suggest_categorical("normalize_type", ["tree", "forest"])
        param["rate_drop"] = trial.suggest_float("rate_drop", 1e-8, 1.0, log=True)
        param["skip_drop"] = trial.suggest_float("skip_drop", 1e-8, 1.0, log=True)

    bst = xgboost.train(param, dtrain)
    preds = bst.predict(dvalid)
    pred_labels = np.rint(preds)
    roc_auc = roc_auc_score(y_test, pred_labels)
    return roc_auc


if __name__ == "__main__":
    study = optuna.create_study(direction="maximize")
    study.optimize(objective, n_trials=100, timeout=600)

    print("Number of finished trials: ", len(study.trials))
    print("Best trial:")
    trial = study.best_trial

    print("  Value: {}".format(trial.value))
    print("  Params: ")
    for key, value in trial.params.items():
        print("    {}: {}".format(key, value))
```
